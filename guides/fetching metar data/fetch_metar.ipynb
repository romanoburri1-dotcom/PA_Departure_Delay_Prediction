Diese Notebook dient dazu, aufzuzeigen, wie METAR-Daten von OpenSky gedownloadet werden können und anschliessend als CSV abgespeichert werden können.

Nötige Libraries importieren

```python
from traffic.data.weather import metar
import pandas as pd
pd.set_option("display.max_columns", None) # 'None' to show all columns.
```

Airport Code (LSZH) eingeben und anschliessend Start- und Enddatum + Zeit eingeben. Unser Dataset von FZAG geht vom 01.01.2023 bis 14.09.2025, deswegen auch genau diese Werte hier im Skript. Es ist möglich, dass es hier zu einem Fehler kommt, wenn bei OpenSky zu viele Requests gleichzeitig ankommen. Lösung: Entweder Rohdaten im Ordner data/raw übernehmen oder den Zeitraum von 2023-2025 auf mehrere CSVs aufteilen. 

```python
# Specify ICAO code of airport, example uses Zurich Airport (LSZH)
airport_code = "LSZH"  

# This query fetches METAR data for a specific airport, identified by its
# ICAO code over a time period defined by start and stop parameters
m = metar.METAR
df_metar = m(airport_code).get(
    start="2023-01-01 00:00", # starting timestamp for the query
    stop="2025-09-14 23:59" # end timestamp for the query
    )
```

Zum Schluss das Dataframe als CSV abspeichern und optional die ersten paar Zeilen anzeigen.

```python
df_metar.to_csv("metar_lszh_2023-2025.csv", index=False)
# Display the first few rows of the fetched METAR data
df_metar.head()
```
