```python
import pandas as pd

def clean_data(df):
    # Drop rows with missing data in columns: 'aircraft_category_icao_code', 'aircraft_type_icao_code', 'runway_concept', 'actual_in_block_time_utc' and 'scheduled_in_block_time_utc'
    df = df.dropna(subset=['aircraft_category_icao_code', 'aircraft_type_icao_code', 'runway_concept', 'actual_in_block_time_utc', 'scheduled_in_block_time_utc'])
    # Filter rows based on column: 'flight_is_general_aviation'
    df = df[df['flight_is_general_aviation'] == False]
    #Filter out duplicates based on off block time and aircraft registration
    df = df.drop_duplicates(subset=['actual_in_block_time_utc', 'aircraft_registration'])
    # Derive column 'Airline' from column: 'flight_identifier'
    # Transform based on the following examples:
    #    flight_identifier    Output
    # 1: "LX016"           => "LX"
    df.insert(1, "Airline", df["flight_identifier"].str[:2])

    return df

# Loaded variable 'df' from URI: c:\PA\data\raw\FZAG_DATA_departures.csv
df = pd.read_csv("../data/raw/FZAG_DATA_arrivals.csv")
df_clean = clean_data(df.copy())

output_path = "../data/processed/clean_FZAG_DATA_arrivals.csv"
#Save cleaned dataframe to CSV
df_clean.to_csv(output_path, index=False)

print(f" Cleaned file saved to: {output_path}")
```
